```python
import numpy as np
import json
import matplotlib.pyplot as plt
import itertools
```

```python
with open('scores.json') as f:
    sas = json.load(f)
```

```python
crps = list(itertools.chain.from_iterable(subdct['PWL']['CRPS'] for case, subdct in sas.items()))
lr = list(itertools.chain.from_iterable(subdct['PWL']['Likelihood Ratio'] for case, subdct in sas.items()))

lr = np.array(lr)[np.argsort(crps)]
crps = np.sort(crps)
```

```python
fig, ax = plt.subplots()

ax.scatter(np.arange(len(crps)), crps, color='C3', s=3, lw=0)
ax.scatter(np.arange(len(crps)), lr, color='C0', s=3, lw=0)
ax.set(xscale='linear', yscale='log', xlim=(0, 530))
ax.grid()
ax.axhline(0.05, color='k')
```
